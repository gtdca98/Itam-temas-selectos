**Filtrar las categorias relevantes** identificadas en el archivo `types_types_s.out` del dropbox. Estas llevan una `K` (de keep) al final. El archivo mencionado fue revisado manualmente para resaltar tipos históricamente relevantes. La elección de tipos es sujeta ampliarse/reducirse desde el archivo.

```python
dropbox = '/Users/jigr/Dropbox/Catalogos/'
type_cat = dropbox + 'types_types_s.out'

dbtypes = list()
for line in open(type_cat):
    if line.rstrip()[-1]=='K':
        dbtypes.append(line[0:-2].split())
        
import pandas as pd
types = pd.DataFrame(dbtypes)
types.rename(columns={0:'count',1:'type'}, inplace=True)

print types
```

Importar _Mapping-based Types_ (MbT):

```python
dbpedia = '/Users/jigr/dbpedia/'
type_file = dbpedia + 'instance_types_en_uris_es.nt'
bytype = pd.read_table(type_file, header=None, delimiter=' ', skiprows=1)
bytype.rename(columns={0:'item',2:'type'}, inplace=True)
bytype.drop([1, 3], axis=1, inplace=True)
```

Filtrar MbT con catálogo de tipos relevantes:

```python
items_lite = pd.merge(types, bytype, how='inner', on='type')
items_lite.drop('count', 1, inplace=True)
items_lite.sort('item', inplace=True)
items_lite.count()
```

```python
items_lite[50000:50500]
```

Aquí un prototipo de la función anónima que se necesita para extraer términos clave de los URL's:

```python
map(lambda x: x.split('/')[-1].translate(None,'>'), 
    '<los/términos/clave> <siempre/vienen> <en/la/posición/final/de/un/url>'.split())
```

Se extrae la información de los URL's a traves de una función anónima que identifica y retiene sólo la última posición de la cadena delimitada por diagonales. Esta versión transformada de los MbT permite identificar de forma sencilla qué pares _tipo-ítem_ se repiten:

```python
items_lite = items_lite.applymap(lambda x: x.split('/')[-1].translate(None,'>'))
items_lite.drop_duplicates(cols=['item'], inplace=True)
items_lite.count()
```

**Tema pendiente**: Si no se eliminan duplicados del ítem, estos se propagan en la creación de nodos y relaciones. Pero plantea una pregunta importante: ¿qué tipo es el que hay que conservar para cada item? Por ejemplo, David Beckham: ¿futbolista, atleta, persona, personalidad? ¿Y si se conservan todos? Lo óptimo parece ser no perder información de los tipos introduciéndolos como atributos de los nodos.

Importar _Mapping-based Properties_ (MbP): Este archivo está más dificil porque no se puede importar como delimitado por espacios. Aquí lo que sucede si se intenta de dicha forma:

```python
prop_file = dbpedia + 'mappingbased_properties_en_uris_es.nt'
byprop = pd.read_table(prop_file, header=None, delimiter=' ', skiprows=1, error_bad_lines=False)
```

La razón es que la columna con el valor de propiedad tiene formato variable: en ocasiones es un URL entre <> pero en otras son valores entre comillas. Se resuelve condicionando la lectura de la siguiente forma:

```python
aux = list()
for line in open(prop_file):
    if line.rstrip()[-5]!="@":
        aux.append(map(lambda x: x.split('/')[-1].translate(None,'>'), line.split()))
    else:
        x = line.split()
        aux.append([x[0].split('/')[-1].translate(None,'>'), 
                    x[1].split('/')[-1].translate(None,'>'),
#                    x[2].split('"')[1]])
                    '_'.join(x[2:len(x)])])

byprop = pd.DataFrame(aux)
byprop.rename(columns={0:'item', 1:'prop', 2:'value'}, inplace=True)
byprop.drop(3, axis=1, inplace=True)
```

En este paso nos deshacemos de propiedades que relacionan ítems ajenos a la base filtrada `items_lite`. Este proceso es fundamental para la carga de la información (es sencillo cargar relaciones cuando ambos nodos ya existen).

```python
props_lite = pd.merge(pd.merge(items_lite[['item']], byprop, how='inner', on='item'), 
                      items_lite[['item']], how='inner', left_on='value', right_on='item')
```

```python
props_lite.drop(props_lite.columns[3], axis=1, inplace=True)
props_lite.rename(columns={'item_x':'item'}, inplace=True)
```

¡El doble merge elimina 5/6 de las relaciones! 
**Tema pendiente:** No se están cuidando las propiedades que apuntan a valores estilo "..."@es_. También sería útil tener un paso de normalización para no deshacernos de términos donde falla el filtrado de relaciones debido a alguna variante de escritura.
```python
gdb_nodes.rename(columns={0:'item'}, inplace=True)
gdb_nodes.drop(gdb_nodes.columns[0], axis=1, inplace=True)
gdb_nodes['item_id'] = pd.DataFrame(range(0, gdb_nodes.count()['item']))
gdb_nodes.head()
```
